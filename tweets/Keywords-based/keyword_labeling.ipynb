```python
def check_aspect(text, aspect):
    for keyword in keywords_dict[aspect]:
        if keyword in text:
            return 1
    return 0
```

```python
df_tweets = df_tweets.loc[:, ['ID', 'Tweets', 'clean_tweets']]
df_tweets
```

```python
for aspect in keywords_dict.keys():
    df_tweets[aspect] = df_tweets['clean_tweets'].apply(lambda x: check_aspect(x, aspect))
```

```python
df_tweets
```

```python
# Remove tweets without any aspect
df_tweets = df_tweets.loc[(df_tweets[['social', 'historical', 'dehumanization', 'accusation', 'attack', 'loyalty', 'threat']] != 0).any(axis=1)]
df_tweets = df_tweets.reset_index(drop=True)
df_tweets
```

```python
# Identifying the number of tweets for each label
import seaborn as sns
import matplotlib.pyplot as plt

# Getting labels
labels = list(df_tweets.columns.values)[3:]

sns.set(font_scale=1)
plt.figure(figsize=(20, 8))

# Combining the labels and values into a DataFrame
label_counts = pd.DataFrame({'labels': labels, 'counts': df_tweets.iloc[:, 3:].sum().values})

# Plotting the bar chart
ax = sns.barplot(x='labels', y='counts', data=label_counts)

plt.title('Tweets per label', fontsize=24)
plt.ylabel('Number of tweets', fontsize=14)
plt.xlabel('Label', fontsize=14)

# Adding the text labels
bars = ax.patches
label_values = label_counts['counts'].values
for bar, label in zip(bars, label_values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 2, label, ha='center', va='bottom', fontsize=18)

plt.show()
```
